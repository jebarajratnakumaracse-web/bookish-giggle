```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample data
x = [10, 20, 30, 40, 50]
y = [12, 22, 28, 35, 55]

# -------------------
# STATISTICS
# -------------------

print("Mean:", np.mean(x))
print("Median:", np.median(x))
print("Mode:", stats.mode(x, keepdims=True).mode[0])
print("Variance:", np.var(x))
print("Standard Deviation:", np.std(x))
print("Correlation:", np.corrcoef(x, y)[0, 1])

# -------------------
# PLOTS
# -------------------

plt.figure(figsize=(12,4))

# 1. Histogram
plt.subplot(1,3,1)
plt.hist(x, color='skyblue', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")

# 2. Box Plot
plt.subplot(1,3,2)
plt.boxplot(x)
plt.title("Box Plot")

# 3. Scatter Plot (Correlation)
plt.subplot(1,3,3)
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("x values")
plt.ylabel("y values")

plt.tight_layout()
plt.show()

```
